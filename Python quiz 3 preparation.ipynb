```python
class A():
    def __init__(self,a,b):
        self__name=a
        self__fname=b
    def display(self):
        print("name",self.__sname)                  
        print("age",self.__age)
        
    def change(self,a,b):   #a,b==name,age   they r local variables 
        self.__sname = a              #here change is SETTER FUNCTION
        self.__age = b
ob=A("anie","omer")
```

```python
class Robot(object):
    def __init__(self,a):
        self.__version = 22

    def getVersion(self):
        print(self.__version)

    def setVersion(self, version):
        self.__version = version

obj = Robot(22)
obj.getVersion()
obj.setVersion(23)
obj.getVersion()
print(obj.__version)
```
